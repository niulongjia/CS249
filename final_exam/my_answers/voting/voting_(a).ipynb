```r
#### code from chapter09.R, which reads in the voting records:
library(foreign)
data.dir <-  "./data/"
data.files <- list.files(data.dir, pattern = ".dta" )
rollcall.data <- lapply(data.files,
                        function(f)
                        {
                          read.dta( f, convert.factors = FALSE)
                        })
rollcall.simplified <- function(df)
{
  no.pres <- subset(df, state < 99)
  for(i in 10:ncol(no.pres))
  {
    no.pres[,i] <- ifelse(no.pres[,i] > 6, 0, no.pres[,i])
    no.pres[,i] <- ifelse(no.pres[,i] > 0 & no.pres[,i] < 4, 1, no.pres[,i])
    no.pres[,i] <- ifelse(no.pres[,i] > 1, -1, no.pres[,i])
  }
  return(as.matrix(no.pres[,10:ncol(no.pres)]))
}
rollcall.simple <- lapply(rollcall.data, rollcall.simplified)

```

```r
no.pres <- subset( rollcall.data[[1]], state < 99 )
head(rollcall.simple[[1]],5)
head(no.pres)
# c("100", "200", "328"),
# c("Dem.", "Rep.", "Ind."))
dim(rollcall.simple[[1]])
dim(no.pres)
```

```r
###  simple code to display a voting heatmap:
plots <- list()  # new empty list
par(mfrow=c(1,2))
for (k in 1:12)
{ 
    #k = 12   # for example:  get the last Congress, the 113th
    f = data.files[k]

    VotingMatrix = rollcall.simple[[k]]
    no.pres <- subset( rollcall.data[[k]], state < 99 )
    SenatorNames = no.pres[,9]
    
    jpeg(paste("biclustering_" ,as.character(k) , ".jpg",sep=""))
    heatmap( VotingMatrix,  scale="none",
             labRow = SenatorNames, cexRow=0.01,
             labCol = colnames(VotingMatrix), cexCol=0.01,
             col=c("red","yellow","green"))
             #  vote values are  -1, 0, 1
    #plots[[k]] = plot1
    legend( "topleft", c('no','--','yes'), col=c("red","yellow","green"), lwd=3, cex=0.35, title=f)
    dev.off()
    
}


#multiplot(plotlist = plots, cols = 2)
```
